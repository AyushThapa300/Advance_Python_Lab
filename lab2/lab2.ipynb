Iterator:
An iterator is an object that allows traversal through all the elements of a collection one at a time. It implements two special methods:

iter() – returns the iterator object.

next() – returns the next element and raises StopIteration when no elements remain.

Example: list, tuple, and string iterators.

```python
num=[10,20,30]
it=iter(num)
print(next(it))
print(next(it))
print(next(it))
```

```python
class MyNumbers:
    def __init__(self,limit):
        self.num=0
        self.limit = limit
    def __iter__(self):
        return self
    def __next__(self):
        if self.num < self.limit:
            self.num += 1
            return self.num
        else:
            raise StopIteration
numbers = MyNumbers(5) 

print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
print(next(numbers))
```

```python
class MyNumbers:
    def __init__(self,limit):
        self.num=0
        self.limit = limit
    def __iter__(self):
        return self
    def __next__(self):
        if self.num < self.limit:
            self.num += 1
            return self.num
        else:
            raise StopIteration
numbers = MyNumbers(5) 

for n in numbers:
    print(n)
```

Generator:
A generator is a special type of iterator that is created using a function with the yield keyword. It generates values one at a time on the fly, instead of storing them in memory, making it memory-efficient.

```python
def square_generator(limit):
    for i in range(1,limit+1):
        yield i * i

gen = square_generator(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def square_generator(limit):
    for i in range(1,limit+1):
        yield i 

gen = square_generator(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def square_generator(limit):
    for i in range(1,limit+1):
        yield i * i

for square in square_generator(5):
    print(square)
```

Decorator:
A decorator is a function that modifies or extends the behavior of another function or method without changing its source code. It uses the @decorator_name syntax.

Example uses: logging, authentication, timing, access control.

```python
def my_decorator(func):
    def wrapper():
        print("Function is about to run")
        func()
        print("Function has finished running")
    return wrapper
@my_decorator
def say_hello():
    print("Hello!")

say_hello()        
```

```python
def my_decorator(func):
    def wrapper():
        print("Function is about to run")
        func()
        print("Function has finished running")
    return wrapper

def say_hello():
    print("Hello!")
say_hello = my_decorator(say_hello)
say_hello()        
```

```python
def my_decorator(func):
    def wrapper():
        print("Before calling the function")
        func()
        print("After calling the function")
    return wrapper
@my_decorator
def greet():
    print("Hello")
greet()        
```

Q. Is for loop an iterator?

ans: No, a for loop is not an iterator. A for loop is a looping (control) statement, not an object. It does not store data or implement iterator methods like iter() and next(). Instead, it works with an iterable (such as a list, tuple, or string) by internally creating an iterator from that iterable.

During execution, the for loop repeatedly calls the iterator’s next() method to fetch elements one by one. When there are no more elements, the iterator raises a StopIteration exception, and the loop automatically stops.

Q. Differences between iterator and generator.

ans: i. An iterator is an object that allows traversal through elements of a collection by implementing the iter() and next() methods, and it may require more code to create. A generator is a special type of iterator created using a function with the yield keyword, which automatically handles iteration and state management.

ii.Generators are more memory-efficient because they produce values lazily, one at a time, instead of storing all values in memory.
